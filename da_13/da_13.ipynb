## 1. 要点まとめ

### 1.1 教師なし学習としてのクラスタ分析

- **教師あり学習**: 正解ラベル（広告費に対する売上、A/B の所属など）を与えて学習する
- **教師なし学習**: 正解ラベルがない。データの構造そのものから「似たもの同士」を見つける

クラスタ分析は後者の代表例で、**距離が近い点を同じグループにまとめる**ことで、隠れたパターンを発見します。

### 1.2 k-means 法のアイデア

最もシンプルで有名な手法が **k-means 法** です。次の手順を繰り返します。

1. クラスタ数 $k$ を決める
2. 空間上に $k$ 個の中心点（セントロイド）をランダムに配置
3. 各データ点を**一番近い中心点**のクラスタに割り当てる
4. 各クラスタの**平均位置**に中心点を更新する
5. 中心点が動かなくなるまで 3〜4 を繰り返す

$k$ の値は分析者が決める必要があります（肘法やシルエット係数などの決め方がありますが、本科目ではビジネス的に意味のある数を指定します）。

### 1.3 特徴量の標準化

k-means は **距離**に基づくため、単位やスケールの異なる列をそのまま使うと、値の大きい列に結果が引きずられます。事前に各列を **標準化**（平均 0、標準偏差 1）してから学習するのが基本です。

```python
from sklearn.preprocessing import StandardScaler

X_scaled = StandardScaler().fit_transform(X)
```

### 1.4 scikit-learn で k-means を 3 行で書く

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, n_init=10, random_state=0).fit(X_scaled)
km.labels_           # 各点のクラスタ番号（0, 1, 2, ...）
km.cluster_centers_  # 各クラスタの中心（標準化後の座標）
```

### 1.5 小さな例でクラスタ分析を体感する

2 つの明らかに離れた集団のダミーデータを作り、`k-means` で 2 クラスタに分けてみましょう。

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

rng = np.random.default_rng(0)
g1 = rng.normal(loc=(2, 3), scale=0.5, size=(30, 2))
g2 = rng.normal(loc=(7, 8), scale=0.5, size=(30, 2))
X = np.vstack([g1, g2])

km = KMeans(n_clusters=2, n_init=10, random_state=0).fit(X)
demo = pd.DataFrame(X, columns=['x', 'y'])
demo['cluster'] = km.labels_

sns.scatterplot(x='x', y='y', hue='cluster', data=demo, palette='Set1', s=60)
# クラスタ中心を重ねる
plt.scatter(
    km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
    marker='X', s=200, color='black', label='中心',
)
plt.legend()
plt.title('2 クラスタの自動分類')
plt.show()
````

ラベルを一切与えていないのに、データの「かたまり」を自動で見つけてくれました。これがクラスタ分析の基本的な挙動です。
