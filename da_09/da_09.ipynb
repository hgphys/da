## 1. 要点まとめ

### 1.1 集計とは：単位を決めて束ねる

生データは「行 = 1 件の取引」のような粒度で並んでいます。集計とは、ある列の値でデータをグループに分け、各グループの合計・平均・件数などを求めて**単位を粗くする**操作です。

- カテゴリごとの売上合計
- 月ごとの平均購入額
- 商品別の販売件数

「〜ごとの〜」という形の問いはすべて集計で答えられます。

### 1.2 groupby の基本

```python
df.groupby('列名')['対象列'].集計関数()
```

よく使う集計関数: `sum()`, `mean()`, `count()`, `max()`, `min()`

### 1.3 pivot_table：2 軸のクロス集計

`groupby` が **1 軸**の集計なのに対し、`pivot_table` は **2 軸**（行と列）を同時に指定したクロス集計表を作れます。Excel のピボットテーブルとほぼ同じ感覚です。

```python
df.pivot_table(
    index='行に置く列',
    columns='列に置く列',
    values='集計したい列',
    aggfunc='sum',  # 集計方法（sum, mean, count など）
)
```

同じ生データでも、`index` と `columns` の組み合わせを変えるだけで、まったく違う「報告書の表」が作れます。集計軸を変えることが、インサイトを発見するための最も基本的な技術です。

### 1.4 小さな例で groupby と pivot_table を体感する

5 人の学生の 2 教科の成績表を使って、`groupby` と `pivot_table` の違いを見てみましょう。

````python
import pandas as pd

scores = pd.DataFrame({
    'name':    ['佐藤', '佐藤', '鈴木', '鈴木', '高橋', '高橋', '田中', '田中', '伊藤', '伊藤'],
    'subject': ['英語', '数学', '英語', '数学', '英語', '数学', '英語', '数学', '英語', '数学'],
    'score':   [78, 85, 62, 74, 90, 88, 55, 60, 81, 92],
})
scores
````

#### groupby：教科ごとの平均点（1 軸）

````python
scores.groupby('subject')['score'].mean()
````

#### pivot_table：学生 × 教科のクロス表（2 軸）

````python
scores.pivot_table(index='name', columns='subject', values='score')
````

同じ 10 行のデータから、`groupby` は 2 行の要約を、`pivot_table` は 5 行 × 2 列のクロス表を作りました。問いに応じて切り口を選ぶのが集計の肝です。
