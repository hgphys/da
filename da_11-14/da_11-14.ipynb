## 背景：クライアントからの相談

あるオンラインショップを運営している企業から、顧客 500 名分の利用実績データを受け取りました。担当者からは以下の 4 件の相談がありました。

**相談 A：**
> 全顧客一律のメルマガでは開封率が伸び悩んでいます。訪問頻度と購入額の組み合わせで顧客を自動的にグループ分けし、グループごとに異なる訴求ができるようにしたいです。

**相談 B：**
> 会員ランクは basic / silver / gold の 3 段階ですが、ランクごとに購入額に本当に差があるのかを確かめたいです。差があるなら、ランク昇格のインセンティブ設計に使えます。

**相談 C：**
> 登録してからの期間が長くなるほど、購入額は本当に増えていくのでしょうか。新規顧客がどのくらい経つとどの程度使うようになるのか、数字で予測できると販促計画が立てやすいです。

**相談 D：**
> スマホ中心のお客さんと PC 中心のお客さんで、購入額の傾向が違う気がしています。サイト改修の優先度を決めたいので、両者に統計的に意味のある差があるか判定してほしいです。
