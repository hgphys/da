### 1.3 「描いてから、書く」の思考手順

毎回の授業で繰り返してきたフローをもう一度確認します。

1. **アプローチ選択クイズ** — 分析アプローチの選択トレーニング
2. **ストーリー構築** — 背景から分析ストーリーを議論
3. **要点解説＋骨組み作成** — 本時の要点を解説し、コードの骨組み（コメント行）を確定
4. **伴走型実装** — 骨組みを Python コードに翻訳
5. **サマリー入力** — 演習結果を記入・提出

このフローの本質は「**問いを立て → 設計図を描き → コードに翻訳し → 結論を言葉にする**」というサイクルです。生成 AI が普及した現在でも、最初の「問い」と最後の「結論」は人間の仕事です。

## 2. 本科目で扱わなかった領域

本科目で学んだ 4 つの統計分析手法は、データ分析の広大な領域の入口です。ここでは、本科目の範囲外にある主な領域を整理します。興味のあるテーマがあれば、関連する科目名をキーワードにして学びを広げてください。

### 2.1 データの構造・目的による広がり

本科目では「行 = サンプル、列 = 少数の変数」の表形式データに対して、人がデータの構造を理解・説明するための統計分析を学びました。現実にはもっと多様なデータ構造があり、また機械にデータの構造を学習させて精度の高い出力を得る機械学習というアプローチもあります。両者は対立するものではなく、回帰やクラスタ分析のように道具は共有されています。

| データの構造 | 本科目との違い | 代表的な手法 | 関連する科目名 |
| :-- | :-- | :-- | :-- |
| 多変量データ | 変数が数十〜数千に増える | 重回帰分析、主成分分析、判別分析 | 多変量解析 |
| 時系列データ | 時間方向の並びに意味がある | ARIMA、季節調整、自己相関分析 | 時系列分析 |
| 多変量時系列データ | 複数変数が同時に時間変化 | VAR モデル、動的因子モデル | 時系列分析、計量経済学 |
| グラフ（ネットワーク）データ | 変数間の関係構造が明示的 | 中心性分析、コミュニティ検出 | ネットワーク科学 |
| 画像データ | ピクセルの2次元配列 | 畳み込みニューラルネットワーク (CNN) | 画像処理、コンピュータビジョン |
| テキストデータ | 自然言語の文や文書 | 形態素解析、TF-IDF、トピックモデル | 自然言語処理 (NLP) |
| 音声データ | 時間方向の波形信号 | スペクトログラム、音声認識モデル | 音声情報処理 |
| — | 精度の高い予測・分類が目的 | 決定木、ランダムフォレスト、SVM、ニューラルネットワーク | 機械学習、パターン認識、深層学習 |

### 2.2 分析の前提に関わる重要なトピック

手法以外にも、実務や研究でデータを扱う際に避けて通れないトピックがあります。

| トピック | 内容 | 関連する科目名 |
| :-- | :-- | :-- |
| 因果と相関の区別 | 相関があることと原因であることは別の主張。回帰で傾きが正でも因果は言えない | 統計学、因果推論 |
| サンプルサイズと検出力 | p 値が有意でなくても差がないとは限らない。サンプルが少なすぎると本当の差を見逃す | 統計学、実験計画法 |
| 個人情報とデータ倫理 | 実データには機微情報が含まれる。匿名化・アクセス管理・同意の取得は分析の前提 | 情報倫理、データガバナンス |
| 結果の再現性 | コードとデータを他者が再実行できるように管理すること | 研究倫理、オープンサイエンス |
| 欠損値の処理 | 本科目では「見つける」まで。どう埋めるかには平均値補完、モデル推定など多くの選択肢がある | データクリーニング、前処理 |

### 2.3 「実際に計算できるか」という問題

本科目では 500〜1000 行程度のデータを扱いましたが、現実には数百万〜数億行のデータや、モデルの計算に数時間〜数日かかるケースが当たり前に存在します。「正しい手法を知っている」だけでは不十分で、「現実的な時間で計算を終えられるか」という工学的な問題がつきまといます。

| トピック | 内容 | 関連する科目名 |
| :-- | :-- | :-- |
| 計算の効率 | 同じ計算でもアルゴリズムの選び方で速度が数桁変わる。計算量（O 記法）の理解が基盤になる | アルゴリズムとデータ構造 |
| ハードウェアの活用 | CPU / GPU / メモリの制約をどう活かすか。GPU による並列計算は深層学習の基盤 | コンピュータアーキテクチャ、並列計算 |
| 大規模データの処理 | Pandas では扱いきれない規模のデータを、分散処理（Spark など）やストリーム処理で扱う技術 | 分散システム、データエンジニアリング |
| 外部計算資源の利用 | 手元の PC で計算しきれない場合に、クラウド上の計算資源を使う。AWS / GCP / Azure など | クラウドコンピューティング |

プログラミングやアルゴリズムの知識は、データ分析を「机上の理論」から「実際に動くシステム」に変えるために不可欠です。本科目で学んだ分析の「何を・なぜ」と、計算機科学が扱う「どうやって・どのくらいの速さで」は、両輪として互いを補い合います。
