## 1. 要点まとめ

### 1.1 なぜ型変換が必要か

Pandas で CSV を読み込むと、各列は自動でデータ型が推定されます。しかし次のような場合、数値として読み込めず文字列（`object` 型）になってしまいます。

- 通貨単位が付いている: `¥1,200`
- 桁区切りのカンマが入っている: `12,345`
- 単位が付いている: `3 個`
- 日付が文字列形式: `2024-01-01`

このままでは `mean()` や大小比較ができません。**文字列の余分な部分を取り除き**、**適切な型に変換する**2ステップで計算可能な形に整えます。

### 1.2 Pandas の変換関数

| 関数 | 役割 | 例 |
| :-- | :-- | :-- |
| `df['列'].str.replace(古, 新)` | 文字列の置換（単位やカンマの除去） | `'¥1,200'` → `'1200'` |
| `df['列'].astype(型)` | 列全体の型を強制変換 | `'1200'` → `1200`（int） |
| `pd.to_datetime(df['列'])` | 文字列を日付型に変換 | `'2024-01-01'` → `Timestamp` |

### 1.3 型変換の基本パターン

文字列から数値への変換は、**置換 → 型変換** の順で行います。

```python
df['price'] = df['price'].str.replace('¥', '').str.replace(',', '').astype(int)
```

`str.replace()` は連鎖（チェーン）させることで複数の文字をまとめて除去できます。
