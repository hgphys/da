## 1. 要点まとめ

### 1.1 Web API とは

**API (Application Programming Interface)** は、プログラムから別のサービスの機能やデータにアクセスするための「窓口」です。特に **Web API** は、HTTP というプロトコルで URL にアクセスすることでデータを返してくれる仕組みで、天気・株価・SNS 投稿・政府統計など、世の中の多くのデータは Web API で取得できます。

ブラウザで URL を開くのと同じことを、Python のコードから実行するイメージです。

### 1.2 JSON 形式

Web API が返すデータの多くは **JSON (JavaScript Object Notation)** という形式です。Python の辞書やリストにほぼそのまま対応するため、読み込んですぐに Pandas の DataFrame に変換できます。

```json
[
  {"id": 1, "title": "最初の投稿", "userId": 10},
  {"id": 2, "title": "2番目の投稿", "userId": 12}
]
```

上の JSON は「2行 × 3列の表」とみなせます。

### 1.3 Python で使うライブラリ

| ライブラリ | 役割 |
| :-- | :-- |
| `requests` | URL にアクセスしてデータを取得する（HTTP クライアント） |
| `pandas.read_json()` | JSON 形式のデータを直接 DataFrame に変換する |

今回は両方を使い、Web API → JSON → DataFrame の流れを体験します。
