## 1. 要点まとめ

### 1.1 メッセージごとに選ぶグラフ

グラフは「何を伝えたいか」（メッセージ）によって選びます。代表的な 4 つのパターンは次のとおりです。

| メッセージ | 代表的なグラフ | 典型的な問い |
| :-- | :-- | :-- |
| **比較** | 棒グラフ | カテゴリ間で差があるか？ |
| **推移** | 折れ線グラフ | 時系列で増えているか？減っているか？ |
| **関係** | 散布図 | 2 つの量に相関があるか？ |
| **分布** | ヒストグラム・箱ひげ図 | 値のばらつき方はどうか？ |

### 1.2 seaborn の基本

`seaborn` は `matplotlib` を使いやすくした可視化ライブラリです。DataFrame をそのまま渡して、列名を軸に指定できます。

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='列名', y='列名', data=df)
plt.show()
```

| 関数 | メッセージ |
| :-- | :-- |
| `sns.barplot()` | 比較 |
| `sns.lineplot()` | 推移 |
| `sns.scatterplot()` | 関係 |
| `sns.boxplot()` | 分布（第8回でも登場） |

### 1.3 4 つのグラフを体感する

簡単なダミーデータで、4 つの基本グラフがそれぞれ何を見せる図なのかを体感しましょう。

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

rng = np.random.default_rng(0)

# ダミーデータ：店舗別・月別の売上
demo = pd.DataFrame({
    'shop':  ['A', 'B', 'C', 'D'] * 6,
    'month': sorted(list(range(1, 7)) * 4),
    'sales': rng.integers(50, 200, size=24),
})
demo.head()
````

#### 比較：棒グラフ（店舗間の売上比較）

````python
sns.barplot(x='shop', y='sales', data=demo, estimator='sum', errorbar=None)
plt.title('店舗別 売上合計')
plt.show()
````

#### 推移：折れ線グラフ（月別の売上推移）

````python
sns.lineplot(x='month', y='sales', hue='shop', data=demo, marker='o')
plt.title('店舗別 月次売上推移')
plt.show()
````

#### 関係：散布図（月と売上の関係）

````python
sns.scatterplot(x='month', y='sales', hue='shop', data=demo, s=80)
plt.title('月と売上の散布図')
plt.show()
````

#### 分布：箱ひげ図（店舗ごとのばらつき）

````python
sns.boxplot(x='shop', y='sales', data=demo)
plt.title('店舗ごとの売上分布')
plt.show()
````

同じ 24 行のデータでも、グラフの種類を変えるだけで伝わるメッセージが変わります。伝えたい内容に合ったグラフを選ぶことが、分析結果を正しく届けるための鍵です。
