## 1. 要点まとめ

### 1.1 なぜ「検定」が必要か

グループ A の平均が 4700、グループ B の平均が 5000 だったとします。「B の方が大きい」と言い切ってよいでしょうか？ 実はこの差は、**たまたま選んだサンプルの偏り**（偶然の誤差）で生じた可能性もあります。

統計検定は、「この差はどれくらい珍しいか」を確率で評価し、**偶然では説明できないほど珍しいなら、本当に差があると判断する**ための手続きです。

### 1.2 帰無仮説と p 値

- **帰無仮説 ($H_0$)**: 「2 つのグループの平均に差はない」という前提
- **対立仮説 ($H_1$)**: 「2 つのグループの平均には差がある」
- **p 値**: 帰無仮説が正しいと仮定したときに、今回観測された差以上の差が**偶然だけで**起きる確率

p 値の解釈はシンプルです。

- p < 0.05 → 偶然とは考えにくい → 帰無仮説を **棄却** し、「差がある」と判断
- p ≥ 0.05 → 偶然の範囲内 → 「差があるとは言えない」と判断（＝差がないと証明したわけではない）

この 0.05 を **有意水準** と呼びます。分野によって 0.01 や 0.10 を使うこともありますが、入門では 0.05 を基準にします。

### 1.3 t 検定：2 つのグループの平均を比較する

2 つのグループの平均に差があるかを調べる最も基本的な検定が **t 検定** です。サンプルサイズが異なる場合や、分散が同じとは限らない場合にも使える **Welch の t 検定** を使います。

```python
from scipy import stats

t, p = stats.ttest_ind(group_a, group_b, equal_var=False)
```

戻り値は **t 統計量**（差の大きさを標準誤差で割ったもの）と **p 値** の 2 つです。

### 1.4 小さな例で t 検定を体感する

「クラス 1 と クラス 2 の数学の点数に差があるか」をダミーデータで検定してみましょう。

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import japanize_matplotlib  # 日本語フォント対策（第1回で導入）
from scipy import stats

class1 = [55, 62, 70, 68, 75, 60, 58, 66, 72, 64]
class2 = [68, 74, 80, 72, 78, 71, 69, 76, 82, 75]

print(f'クラス1 平均: {np.mean(class1):.1f}')
print(f'クラス2 平均: {np.mean(class2):.1f}')

t, p = stats.ttest_ind(class1, class2, equal_var=False)
print(f't = {t:.3f}, p = {p:.4f}')

judge = '差がある' if p < 0.05 else '差があるとは言えない'
print(f'→ 有意水準 5% で判定すると「{judge}」')

df_demo = pd.DataFrame({
    'class': ['クラス1'] * len(class1) + ['クラス2'] * len(class2),
    'score': class1 + class2,
})
sns.boxplot(x='class', y='score', data=df_demo)
plt.title('クラス別 数学の点数')
plt.show()
````
