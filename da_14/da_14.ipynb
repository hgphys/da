第12回の t 検定は **2 つのグループ**の平均を比べる手法でした。では、**3 つ以上のグループ**を同時に比べたいときはどうすればよいでしょうか？ 「3 つのうちどれかに差があるか」を 1 回で判定できる手法が **分散分析 (ANOVA)** です。

## 1. 要点まとめ

### 1.1 なぜ t 検定を繰り返してはいけないか

3 つのグループ A, B, C があるとき、A-B, A-C, B-C の 3 ペアに分けて t 検定を繰り返したくなります。しかし、各検定で 5% の誤判定リスク（第一種の過誤）があるため、3 回繰り返すと **全体として誤判定する確率は約 14% まで跳ね上がります**。これを **多重検定の問題**と呼びます。

分散分析は、**1 回の検定で「3 つ以上のグループのどこかに差があるか」を判定する**ことでこの問題を回避します。

### 1.2 分散分析のアイデア：群間分散と群内分散

全データのばらつきを、次の 2 つに分解します。

- **群間分散**: グループ間の平均の差によるばらつき（「本当に差があれば大きい」）
- **群内分散**: 各グループ内の個人差によるばらつき（「ノイズ」に相当）

その比を **F 統計量** と呼びます。

$$
F = \frac{\text{群間分散}}{\text{群内分散}}
$$

F が大きいほど「ノイズに比べて群間の差が目立つ」ことを意味し、帰無仮説（すべてのグループの平均は等しい）が棄却されやすくなります。得られた F 値から **p 値**を計算する流れは t 検定と同じです。

### 1.3 一元配置分散分析（one-way ANOVA）

1 つの要因（例: 店舗レイアウト）について、グループ間で平均に差があるかを調べるのが **一元配置分散分析**です。scipy では 1 行で実行できます。

```python
from scipy import stats

f, p = stats.f_oneway(group_a, group_b, group_c, group_d)
```

### 1.4 分散分析の限界：どこに差があるかまでは分からない

分散分析で p < 0.05 になっても、それは「**どこかに**差がある」ことを示すだけで、A・B・C・D のどのペアに差があるのかまでは分かりません。具体的なペアを特定したい場合は、**Tukey の HSD 検定**などの **多重比較法**を使いますが、本講義では「ANOVA で有意 → ペア比較へ進む」という**流れの理解**に留めます。

### 1.5 小さな例で分散分析を体感する

3 つの指導法による小テストの点数（各 6 名）を比較してみましょう。

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

method_a = [62, 68, 65, 70, 66, 64]
method_b = [72, 75, 78, 74, 73, 76]
method_c = [65, 69, 67, 70, 68, 66]

print(f'A 平均: {np.mean(method_a):.1f}')
print(f'B 平均: {np.mean(method_b):.1f}')
print(f'C 平均: {np.mean(method_c):.1f}')

f, p = stats.f_oneway(method_a, method_b, method_c)
print(f'F = {f:.3f}, p = {p:.4f}')

judge = '少なくとも 1 つに差がある' if p < 0.05 else '有意な差は認められない'
print(f'→ 有意水準 5% で判定すると「{judge}」')

demo = pd.DataFrame({
    'method': ['A']*6 + ['B']*6 + ['C']*6,
    'score': method_a + method_b + method_c,
})
sns.boxplot(x='method', y='score', data=demo)
plt.title('指導法別の小テスト点数')
plt.show()
````
