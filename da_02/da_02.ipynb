## 1. 要点まとめ

### 1.1 データの種類：量的データと質的データ

分析で扱うデータは、大きく 2 つの種類に分かれます。使える分析手法や要約方法がそれぞれ異なるため、まず列ごとに「どちらの種類か」を見分けることが重要です。

| 種類 | 説明 | 例 | 代表的な要約 |
| :-- | :-- | :-- | :-- |
| 量的データ（quantitative） | 数値として大小比較や計算ができるデータ | 価格、販売数、気温、身長 | 平均・中央値・最小最大 |
| 質的データ（qualitative） | カテゴリや名称など、計算ではなく分類に使うデータ | 商品名、カテゴリ、性別、曜日 | 頻度（出現回数） |

Pandas では、量的データはおおむね `int64` / `float64`、質的データは `object`（文字列）として読み込まれます。

### 1.2 データソースの観点

データ分析を始める前には、そのデータが **どこから来たのか**（データソース）を把握しておく必要があります。

- **内部データ**: 自社の販売記録・顧客データベース・アンケート結果など
- **外部データ**: 政府統計、オープンデータ、Web API、SNS など

ソースによって更新頻度・信頼性・利用許諾が異なるため、分析結果の解釈もこれに依存します。

### 1.3 Pandas の要約関数

| 関数 | 役割 | 向いているデータ |
| :-- | :-- | :-- |
| `df.dtypes` | 各列のデータ型を確認 | 全列 |
| `df.describe()` | 平均・標準偏差・最小最大などの要約統計量 | 量的データ |
| `df['列名'].value_counts()` | 各値の出現頻度 | 質的データ |
