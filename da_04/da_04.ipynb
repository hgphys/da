前回までで、CSV や Web API からデータを Python に取り込めるようになりました。しかし実際の業務データには、欠損（空欄）・重複・異常値といった **「汚れ」**がしばしば混ざっています。汚れたまま分析すると、平均や合計が歪んで誤った結論を導きかねません。今回はこの汚れを **特定** する方法を学びます（**除去・補完** の技術は後続の回で扱います）。

## 1. 要点まとめ

### 1.1 データの汚れの三大要素

| 種類 | 内容 | 例 |
| :-- | :-- | :-- |
| 欠損値 (missing) | 値が記録されていない空欄 | 価格の入力漏れ |
| 重複 (duplicate) | 同じ行が複数回登録されている | 二重登録された取引記録 |
| 異常値 (outlier) | 常識的にあり得ない極端な値 | 販売数 9999 個 |

### 1.2 汚れが分析結果に与える影響

- **欠損値**: 欠損の多い列を使った平均や相関は信用できません
- **重複**: 合計値が水増しされ、売上や件数が実際より多く見えます
- **異常値**: 1 件の極端な値だけで平均や標準偏差が大きく引きずられます

### 1.3 Pandas の確認関数

| 関数 | 役割 |
| :-- | :-- |
| `df.isnull().sum()` | 列ごとの欠損件数 |
| `df.duplicated().sum()` | 完全重複行の件数 |
| `df[df['列'] > 値]` | ブール索引で条件に合う行だけを抽出（異常値の抽出にも使える） |

> 補足: SQL 風に `df.query('列 > 値')` と文字列で条件を書く方法もありますが、本科目ではまずブール索引を基本として学びます。
