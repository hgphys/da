実務のデータは、1つの表にすべてがまとまっていることは稀です。販売データ・顧客マスタ・商品マスタ・在庫データなど、バラバラの表を **共通の列（キー）** でつなぎ合わせて初めて分析ができるケースがほとんどです。今回は Pandas の `merge` を使って、複数の表を 1 つに結合する方法を学びます。

## 1. 要点まとめ

### 1.1 なぜ結合が必要か

実務のデータは、1 つの表にすべてがまとまっていることは稀です。たとえば **学生の基本情報**と **試験の成績**は別の表に保管されているのが普通で、両者を組み合わせて初めて「学年別の平均点」などを分析できます。

| students（学生マスタ）の例 | scores（試験結果）の例 |
| :-- | :-- |
| id, name, grade | id, subject, score |

この 2 つを **共通の列（キー）** である `id` でつなぎ合わせる操作が **結合 (merge)** です。

### 1.2 pd.merge の基本

```python
result = pd.merge(left_df, right_df, on='結合キーの列名', how='inner')
```

- `on` で指定した列の値が一致する行同士がくっつきます
- `how` で結合の仕方を指定できます（既定は `inner`）

| `how` の値 | 振る舞い |
| :-- | :-- |
| `'inner'` | 両方の表にキーがある行だけ残す（既定） |
| `'left'` | 左の表は全行残し、右に無ければ `NaN` |
| `'right'` | 右の表は全行残し、左に無ければ `NaN` |
| `'outer'` | どちらかにあれば全行残す |

### 1.3 結合後の確認ポイント

結合後は必ず次の 2 点を確認します。

- **行数**が想定どおりか（予期せぬ増減がないか）
- **欠損**が発生していないか（マスタに存在しないキーがあると `NaN` になる）


### 1.4 小さな例で結合を体感する

学生 4 人の名簿 `students` と、3 人分の試験結果 `scores` があるとします。共通キーは `id` です。`students` の田中さんは試験を受けていないため `scores` には存在しません。

````python
import pandas as pd

students = pd.DataFrame({
    'id':   [1, 2, 3, 4],
    'name': ['佐藤', '鈴木', '高橋', '田中'],
    'grade': [2, 2, 3, 2],
})
students
````

````python
scores = pd.DataFrame({
    'id':      [1, 2, 3],
    'subject': ['英語', '数学', '英語'],
    'score':   [78, 85, 92],
})
scores
````

#### 内部結合 (`how='inner'`)

両方の表に `id` がある行だけを残します。田中さん（id=4）は消えます。

````python
pd.merge(students, scores, on='id', how='inner')
````

#### 左結合 (`how='left'`)

左の表（学生名簿）は全員残し、スコアが無い田中さんの subject・score は `NaN` になります。

````python
pd.merge(students, scores, on='id', how='left')
````

同じデータでも `how` の指定次第で行数と欠損の有無が変わります。**結合後は必ず `shape` と欠損を確認する**習慣をつけましょう。
